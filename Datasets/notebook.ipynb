```python
import pandas as pd
Data_df=pd.read_csv("atlcrime.csv")
Data_df
```

```python
neighborhood = 'Downtown' # Cambiar por el barrio que se quiera consultar
downtown_deaths = Data_df[Data_df['neighborhood'] == neighborhood] # Filtrar por barrio seleccionado y guardar en variable downtown_deaths
total_deaths_downtown = downtown_deaths['neighborhood'].value_counts()[neighborhood] # Sumar los valores de la columna neighborhood y guardar en variable total_deaths_downtown
print("Sumatorio de muertes en Downtown:", total_deaths_downtown) # Mostrar por pantalla el sumatorio de muertes en el barrio seleccionado
downtown_deaths # Mostrar tabla con los datos filtrados
```
